```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('pictures.csv')
df.head()
```

```python
df = df.drop(columns = ['Unnamed: 0', 'Metascore','Movie_runtime'])
```

```python
df
```

```python
def list_make(columns):
    g=[]
    for i in df[columns]:
        t = i.split(', ')
        g.append(t)
    df[columns] = g  
```

```python
list_make(columns = 'Stars')
```

```python
df.sample(10)
```

```python
list_make(columns = 'Movie_director')
```

```python
list_make(columns = 'Genre')
```

```python
df.info()
```

```python
df['Stars'] = df['Stars'].apply(lambda x: [i.replace(' ', '') for i in x])
```

```python
df['Movie_director'] = df['Movie_director'].apply(lambda x: [i.replace(' ', '') for i in x])
df['Genre'] = df['Genre'].apply(lambda x: [i.replace(' ', '') for i in x])
```

```python
df.head()
```

```python
df['Description'] = df['Description'].apply(lambda x:x.split())
```

```python
df['Title'] = df['Title'].apply(lambda x:x.split())
```

```python
df
```

```python
df['tags'] = df['Description'] + df['Movie_director'] + df['Stars'] + df['Genre'] + df['Title']
```

```python
df.head()
```

```python

sf = df[['Image','Title','tags','Rating','Year_of_Release']].head(1750)
```

```python
movie_name = []
for i in sf['Title']:
    j = ' '.join(i)
    movie_name.append(j)
    
```

```python
sf['Title'] = movie_name
```

```python
sf['tags'] = sf['tags'].apply(lambda x:" ".join(x))
```

```python
sf.head()
```

```python
sf['tags'] = sf['tags'].apply(lambda x:x.lower())
```

```python
sf.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
cv = CountVectorizer(max_features = 5000, stop_words = 'english')
tfidf=TfidfVectorizer(max_features = 5000)
```

```python
vectors = cv.fit_transform(sf['tags']).toarray()
vec = tfidf.fit_transform(sf['tags']).toarray()
```

```python
tfidf.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
similarity_tfidf = cosine_similarity(vec)
```

```python
similarity_tfidf.shape
```

```python
def recommend(movie,x):
    index = sf[sf['Title'] == movie].index[0]
    distances = sorted(list(enumerate(x[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(sf.iloc[i[0]].Title)
```

```python
sf.head(20)
```

```python
recommend("The Wolf of Wall Street",similarity_tfidf)
```

```python
recommend("The Wolf of Wall Street",similarity)
```
