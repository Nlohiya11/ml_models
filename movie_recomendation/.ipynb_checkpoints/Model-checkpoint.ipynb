```python
 def page_num(website):
    headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
    webpage = requests.get(website, headers =headers).text
    len(webpage)
    soup = BeautifulSoup(webpage, 'html.parser')
    page = soup.find('span', class_ = 'pagination-range').text.strip().split(' - ')[1].split(' of ')
    movies_on_page_1 = int(page[0])
    try:
        total_num = int(page[1].split(',')[0])*1000 + int(page[1].split(',')[1])
    except:
        total_num = int(page[1])
    total = int(total_num/movies_on_page_1)
    total_pa = total_num/movies_on_page_1
    if total_pa>total:
        total_pages = total+1
    else:
        total_pages = total
    return total_pages
```

```python
def every_movie(website,pages):
    headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
    for i in range(1,pages+1):
        webpage= ""
        string = requests.get(website.format(i), headers =headers).text
        webpage = string + webpage
        soup = BeautifulSoup(webpage, 'html.parser')
        movies = soup.find_all('div', class_ = 'lister-item mode-detail')
        for info in movies:
            i = info.find('h3', class_='lister-item-header').text.strip()
            name.append(i.split('\n')[1])
            try:
                year.append(i.split('\n')[2].split('(')[1].replace('(','').replace(')',''))
            except:
                year.append(np.nan)
            try:
                rating.append(float(info.find('span', class_='ipl-rating-star__rating').text.strip()))
            except:
                rating.append(np.nan)
            try:
                runtime.append(int(info.find('span', class_='runtime').text.strip().split(' ')[0]))
            except:
                runtime.append(np.nan)
            try:
                certificate.append(info.find('span', class_='certificate').text.strip().split(' ')[0])
            except:
                certificate.append(np.nan)
            try:
                genre.append(info.find('span', class_='genre').text.strip())
            except:
                genre.append(np.nan)

        sit = soup.find_all('p', class_='text-muted text-small')
        directors=[]
        cast=[]
        for i in sit:
            try:
                r = i.text.strip()
                index_dir = r.find('Director')
                index_star = r.find('Stars')
                if index_star != -1:
                    try:
                        star = r.split('Stars:\n')[1].split('\n| ')[0]
                        cast.append(star.replace('\n',''))
                    except:
                        rq=np.nan
                else:
                    cast.append(np.nan)
                if index_dir != -1:
                    try:
                        director = r.split('Director:\n')[1].split('\n| ')[0]
                        directors.append(director)
                    except:
                        try:
                            director = r.split('Directors:\n')[1].split('\n| ')[0]
                            director = director.replace('\n','')
                            directors.append(director)
                        except:
                            q=0
                else:
                    directors.append(np.nan)
            except:
                q=0
        length_dir = int(len(directors)/3)
        length_star = int(len(cast)/3)
        for i in range(0,length_dir):
            movie_director.append(directors[1+3*i])
        for i in range(0,length_dir):
            stars.append(cast[1+3*i])
        print(len(name),len(year),len(rating),len(runtime),len(certificate),len(genre),len(movie_director),len(stars))
```

```python
pages = page_num('https://www.imdb.com/list/ls098483639/?sort=list_order,asc&st_dt=&mode=detail&page=1')
every_movie('https://www.imdb.com/list/ls098483639/?sort=list_order,asc&st_dt=&mode=detail&page={}',pages)
```
