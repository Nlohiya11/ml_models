```python
scaler = StandardScaler()
reg = LinearRegression()
pipe = make_pipeline(column_trans,scaler,reg)
```

```python
pipe.fit(x_train,y_train)
```

```python
Y_pred_lr=pipe.predict(x_test)

r2_score(y_test,Y_pred_lr)
```
